```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
import requests
```

Creating & testing the tool DuckDuckGoSearchRun

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()
results = search_tool.invoke("news in India today? summarize in 5 lines")
print(results)
```

Creating a Wheather Tool

```python
@tool
def get_weather_data(city: str) -> str:
  """
  This function fetches the current weather data for a given city
  """
  url = f'https://api.weatherstack.com/current?access_key= {WEATHERSTACK_API_KEY} & query={city}'

  response = requests.get(url)

  return response.json()
```

Creating the LLM & testing it

```python
llm = ChatOpenAI()
llm.invoke("hi")
```

Creating Agent & Agent Executor with a help of pre build prompt from langchain hub

```python

from langchain_classic.agents import create_react_agent,AgentExecutor
from langchain_classic import hub
```

```python
prompt = hub.pull("hwchase17/react")
```

```python
agent = create_react_agent(
    llm=llm,
    tools=[search_tool,get_weather_data],
    tools=[search_tool,],
    prompt=prompt
)
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool,get_weather_data],
    verbose=True
)
```

lets see how agent executor runs a query with verbose set to true

```python
response = agent_executor.invoke({"input": "show me 3 ways to reach goa from Mumbai"})
print(response)
response["output"]
```

```python
response = agent_executor.invoke({"input": "Find the capital of Gujarat, then find it's current weather condition"})
print(response)
response["output"]
```
